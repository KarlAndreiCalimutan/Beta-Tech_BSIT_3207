# **Consumer Price Indexes between 2001-2020**

*   The following data presents the information regarding to the Consumer Price Index that happens in 20 years. This dataset helps the researchers to discuss the inflation rate happening per year in accordance to the phenomenon being experienced.


```python
#Dataframe
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/CPI%20per%20year.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all the values stored in the dataframe/csv
Year = df["Year"]
CPI = df["CPI"]

#changing the size of visualization
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#creating the plot
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("\n Consumer Price Index (2001-2020)", fontweight="bold", fontsize=20)
plt.xlabel("Reference Date", fontsize=15)
plt.ylabel("CPI", fontsize=15)
plt.plot(Year, CPI, 'o-', c= '#333399', lw=2.4)
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.legend(labels=['CPI'])
plt.show()
```

# **Interpretation**

This line graph here represents the CPI or the Cosumer Price Index in the past 20 years. The x-axis represents the years in between 2001-2020 on the other hand the y-axis represents the CPI number of data. The diagram clearly demonstrate that from year 2001 upto 2008 it is on a steady rise except for that 1 time in 2009 that it dropped for a while. Since then, it becomes steadily inclined again and had a minimal drop in 2016 and it rise again up until 2020.

# **Analysis**

Consumer price index or CPI show the inflation rate of the certain year. In this CPI the base year is 2020 we can see here since 2001 it has been steady rise showing consistent until that drop in 2009 which may have been caused the financial crisis on that year, but besides that it continued to steadily rise again and stand where it is right now.

# CORRELATION MATRIX OF FLOOD DEATHS IN SOUTH EAST ASIAN COUNTRY



*   The Figure below help us analyze how the dependent variable is affected by the changes in the independent variable. Total affected was used as our dependent variable since this is what we will be anticipating for the following years, which is included in our independent variables together with the Number of Injured, Homelesness, CPI, Dis Mag Value, Total affected and Total Deaths.
*   Correlation Matrix is used to demonstrate and present the correlation coefficients of the mentioned variables.






```python
f = plt.figure(tight_layout=True)
f.set_figwidth(10)
f.set_figheight(7)

corMatrix = df.corr().round(1)
sns.heatmap(corMatrix, annot=True, cmap="flare", square=True)
plt.title("Correlation Flood Deaths in South East Asian Country", fontweight= "bold")
plt.show()
```

# **Line Regression Analysis**

### YEAR VS Consumer Price Index (CPI): 1
The calculated correlation between CPI and Year is 1, indicating a relationship to each other, which means that every passing year, the Consumer Price Index increases.


```python
#changing the size of visualization
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)
#Linear Regression
sns.lmplot(x="Year", y="CPI", data=df)
plt.title("\n Relationship Between Year and CPI", fontweight="bold")
plt.show()
```
