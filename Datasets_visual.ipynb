# **Total Flood Damages in ASEAN Countries in between 2001-2020.**


*   The following data provides an information about the overall damages caused by the flood in each country. This dataset will helps the researchers to determine the diifferences of the overall damages per country.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#font
plt.rcParams["font.family"] = "serif"
```

```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20damages%20%24%20in%20over%2020%20years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all values from the dataframe/csv
Country = df["Country"]
Total_damage_adjusted= df["Total damage adjusted in $"]

#size of the bar graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#creating visualization
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Asian Countries", fontsize= 15)
plt.xlabel("Damages Records", fontsize= 15)
plt.title("\n Flood Damages in South East Asia (2001-2020)", fontweight="bold", fontsize=20)
colors = ['#ff7c43', '#ffa600', '#ff7c43', '#f95d6a', '#d45087','#a05195', '#665191', '#2f4b7c', '#003f5c']
plt.barh(Country, Total_damage_adjusted, color = colors )
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('FloodDamages.tiff', transparent=True, dpi=300)
plt.show()
```

# **Interpretation**
The bar graph presented is entitled "Flood Damages in Asean (2001-2020)", which showcases the total amount of damages caused by the flood. As shown in the diagram, the total damages varies according to its countries. The y-axis is composed of the ASEAN countries; on the other hand, the x-axis represents the number of total damages. As shown in the diagram it depicts the 9 countries with corresponding records. The diagram shows that Thailand is the most damaged country with the total damages record of 52,863,787. In contrast, Timor-Leste is the least damaged country with the total damages record of 13,636. Furthermore, Vietnam is the second most damaged country with 18,129,292 total records. Next to it are Indonesia, Malaysia, Philippines and Cambodia. There
are two countries with least equal total number damages: the Lao People and Myanmar.

In the graph we are shown the country with the highest flood damage and the one with the lowest flood damage we have their Thailand as the most damaged country and Timor Leste with the lowest but we will not be able to compare them because Timor Leste is a small country so we will use Myanmar. With relation to the topography of both countries they are both in the ring of fire but if we go and look into a map we will see that in Thailand, typhoons have a more open path into Thailand while on Myanmar there are several bodies of land that will break the velocity or the momentum of the typhoon, but above all Myanmar has more trees and less slopes than Indonesia. ## **ASIA Countries' Flood Death Rate in 20 Years.**


*   The following data provides an information regarding to the total number of death rate that happens in 20 years. This datasets help the researchers in identifying what country suffers the most and the least in terms of death rate in accordance to Flood phenomenon.


```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20Deaths%20in%2020%20Years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#Visualization
#getting all values from the dataframe/csv
Country = df["Country"]
Deaths = df["Deaths"]

#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Creating plots
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("\n Flood Death Rate in Asia Country", fontweight="bold", fontsize=20)
plt.xlabel("Country", fontsize=15)
plt.ylabel("Death Count", fontsize=15)
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.plot(Country, Deaths, 'o-', lw=2.4, c='#333399')
plt.legend(labels=['Death Rate'])
plt.show()
```

# **Interpretation**
The Line graph illustrates the "Flood Death Rate in Asean Country". As shown in the diagram, the x-axis labels the countries in Southeast Asia while on the other hand the y-axis are the number of total death counts. As indicated in the diagram there are certain various data points being pointed in accordance to its country. Vietnam, Thailand and Indonesia has an increasing slope as to the number of total deaths. In contrast, Cambodia, Lao, Malaysia, Myanmar and Philippines has a minimal gap to one another.


# **Analysis**

In this graph here shows that Vietnam have the most highest deaths caused by flood this means that flash floods or heavy currented floods are occuring in Vietnam be advised that these are deaths due to flood and flood only not including landslides or of the sort. This graph shows that a sense of urgency in times of calamity must be practiced given that deaths are apparent citizen must always be prepared in the first sign of disaster to prevent unwanted results.


*    Based on the study of National Library of Medicine (2012) which entitled "Floods in Southeast Asia: A health priority" which discuss that, There is the potential for direct harm and death during the flood's initial stages. The majority of deaths directly linked to the commencement of a flood are caused by drowning, which commonly occurs when people attempt to drive through high, fast-moving water in their vehicles or are swept away into flow of water floods while attempting to assist others. Other causes of mortality include being struck by objects in fast moving water, heart attacks, electrocution, and being buried in mud or infrastructure collapse.

# **South East Asia Countries' total Flood Subtype Incident in 20 years.**

*   The following data provides an information to the total number of Flood Subtype that happens in the ASEAN countries in 20 years. This datasets helps the researchers to visualize what country suffered the most and the least in terms of flood subtype.

```python
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)
Country = df['Country'].values
x = np.arange(len(Country))
w = 0.3
plt.title("\n Floodsubtype for Each Country", fontweight="bold", fontsize=20)
plt.bar(x-w, df['FLASHFLOOD'].values, width=w, label='FLASHFLOOD')
plt.bar(x, df['RIVERINE FLOOD'].values, width=w, label='RIVERINE FLOOD')
plt.bar(x+w, df['COASTAL FLOOD'].values, width=w, label='COASTAL FLOOD')
plt.xticks(x, Country)
plt.ylim([0,40])
plt.tight_layout()
plt.xlabel('X label')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, ncol=5)
plt.show()
```

This Bar graph represents the Flood subtypes that occured in each country in the span of 20 years. As shown in the diagram, the number of occurence varies accorrding to its subtype of flood. The x-axis labes the ASEAN countries while on the y-axis are the total number of incident. The bar graphs are categorized 
by their colors. The blue bars indicate Flash Floods; the orange bar is Riverine
Flood; and lastly, the green bar indicate Coastal Flood. Based on the graph, most of the subtype of flood that occurs in 20 years is Riverine Flood, the country that suffers the most is the Philippines. In contrast, the least subtype of flood that occurs in 20 years is Coastal Flood in which majority of it came from Vietnam and few from Philippines. # **Consumer Price Indexes between 2001-2020**

*   The following data presents the information regarding to the Consumer Price Index that happens in 20 years. This dataset helps the researchers to discuss the inflation rate happening per year in accordance to the phenomenon being experienced.


```python
#Dataframe
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/CPI%20per%20year.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python

#getting all the values stored in the dataframe/csv
Year = df["Year"]
CPI = df["CPI"]

#changing the size of visualization
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#creating the plot
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("\n Consumer Price Index (2001-2020)", fontweight="bold", fontsize=20)
plt.xlabel("Reference Date", fontsize=15)
plt.ylabel("CPI", fontsize=15)
plt.plot(Year, CPI, 'o-', c= '#333399', lw=2.4)
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.legend(labels=['CPI'])
plt.show()
```

# **Interpretation**

This line graph here represents the CPI or the Cosumer Price Index in the past 20 years. The x-axis represents the years in between 2001-2020 on the other hand the y-axis represents the CPI number of data. The diagram clearly demonstrate that from year 2001 upto 2008 it is on a steady rise except for that 1 time in 2009 that it dropped for a while. Since then, it becomes steadily inclined again and had a minimal drop in 2016 and it rise again up until 2020.

# **Analysis**

CPI can show the inflation rate of the certain year. In this CPI the base year is 2022 we can see here since 2002 it has been steady rise showing good promise until that drop in 2009 which may have been caused the financial crisis on that year, but besides that it continued to steadily rise again and stand where it is right now.
