## **Datasets of Flood Impact in Southeast Asia between 2001 to 2020**

*   The following data provides an information on the total Flood Impact in Southeast Asia between 2001 to 2020 adjusted into Dollar currency. The number of total records in this dataframe are divided into three: The total Damages adjusted which is the total number of damaged to property. The total death rate which is the total number of deaths. Lastly, Subtype of flood that occurs throughout the year.  




```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#font
plt.rcParams["font.family"] = "serif"
```

```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20damages%20%24%20in%20over%2020%20years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all values from the dataframe/csv
Country = df["Country"]
Total_damage_adjusted= df["Total damage adjusted in $"]

#size of the bar graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#creating visualization
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Asian Countries", fontsize= 15)
plt.xlabel("Damages Records", fontsize= 15)
plt.title("\n Flood Damages in ASEAN (2001-2020)", fontweight="bold", fontsize=20)
colors = ['#ff7c43', '#ffa600', '#ff7c43', '#f95d6a', '#d45087','#a05195', '#665191', '#2f4b7c', '#003f5c']
plt.barh(Country, Total_damage_adjusted, color = colors )
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('FloodDamages.tiff', transparent=True, dpi=300)
plt.show()
```

# **Interpretation**
The bar graph presented is entitled "Flood Damages in Asean (2001-2020)", which showcases the total amount of damages caused by the flood. As shown in the diagram, the total damages varies according to its countries. The y-axis is composed of the ASEAN countries; on the other hand, the x-axis represents the number of total damages. As shown in the diagram it depicts the 9 countries with corresponding records. The diagram shows that Thailand is the most damaged country with the total damages record of 52,863,787. In contrast, Timor-Leste is the least damaged country with the total damages record of 13,636. Furthermore, Vietnam is the second most damaged country with 18,129,292 total records. Next to it are Indonesia, Malaysia, Philippines and Cambodia. There
are two countries with least equal total number damages: the Lao People and Myanmar.

```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20Deaths%20in%2020%20Years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#Dataframe
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/FLOOD-SUBTYPES.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)
Country = df['Country'].values
x = np.arange(len(Country))
w = 0.3
plt.title("\n Floodsubtype for Each Country", fontweight="bold", fontsize=20)
plt.bar(x-w, df['FLASHFLOOD'].values, width=w, label='FLASHFLOOD')
plt.bar(x, df['RIVERINE FLOOD'].values, width=w, label='RIVERINE FLOOD')
plt.bar(x+w, df['COASTAL FLOOD'].values, width=w, label='COASTAL FLOOD')
plt.xticks(x, Country)
plt.ylim([0,40])
plt.tight_layout()
plt.xlabel('X label')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, ncol=5)
plt.show()
```

# **Interpretation**

This Bar graph represents the Flood subtypes that occured in each country in the span of 20 years. As shown in the diagram, the number of occurence varies accorrding to its subtype of flood. The x-axis labes the ASEAN countries while on the y-axis are the total number of incident. The bar graphs are categorized 
by their colors. The blue bars indicate Flash Floods; the orange bar is Riverine
Flood; and lastly, the green bar indicate Coastal Flood. Based on the graph, most of the subtype of flood that occurs in 20 years is Riverine Flood, the country that suffers the most is the Philippines. In contrast, the least subtype of flood that occurs in 20 years is Coastal Flood in which majority of it came from Vietnam and few from Philippines.
```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Correlation%20matrix.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
df.describe()
```

```python
#Overview of Relationships
survived_summary = df.groupby("TotalDeaths")
survived_summary.mean().reset_index()
```

```python
#Correlations
pd.DataFrame(abs(df.corr()['TotalDeaths']).sort_values(ascending = False))
```

**With correlation to the change in TotalDeaths Noaffected or the number of affected is the closest feature to it it is the most correlated.**

```python
## get the most important variables. 
corr = df.corr()**2
pd.DataFrame(corr.TotalDeaths.sort_values(ascending=False))
```

**If squared the correlation feature also gives on positive correlation it also amplifies the relationships **

```python
#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(10)
f.set_figheight(7)

#Correlation and Heatmap Matrix
corMatrix = df.corr().round(1)
sns.heatmap(corMatrix, annot=True, cmap="flare", square=True)
plt.title("Correlation Flood Deaths in Asean Country", fontweight= "bold")
plt.show()
```
