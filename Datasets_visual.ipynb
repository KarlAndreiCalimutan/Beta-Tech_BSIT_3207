## **Datasets of Flood Damages in Southeast Asia between 2000 to 2020**


```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

#font
plt.rcParams["font.family"] = "serif"
```

```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20damages%20%24%20in%20over%2020%20years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#getting all values from the dataframe/csv
Country = df["Country"]
Total_damage_adjusted= df["Total damage adjusted in $"]

#size of the bar graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#creating visualization
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Asian Countries", fontsize= 15)
plt.xlabel("Damages Records", fontsize= 15)
plt.title("\n Flood Damages in ASEAN (2000-2002))", fontweight="bold", fontsize=20)
colors = ['#003f5c', '#58508d', '#2f4b7c', '#665191',
          '#a05195', '#d45087', '#f95d6a', '#ff7c43', '#ffa600']
plt.barh(Country, Total_damage_adjusted, color = colors )
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('FloodDamages.tiff', transparent=True, dpi=300)
plt.show()
```

# **Interpretation**
The bar graph presented is entitled "Flood Damages in Asean (2000-2002)", which showcases the total amount of damages caused by the flood. As shown in the diagram, the total damages varies according to its countries. The y-axis is composed of the ASEAN countries; on the other hand, the x-axis represents the number of total damages. As shown in the diagram it depicts the 9 countries with corresponding records. The diagram shows that Thailand is the most damaged country with the total damages record of 52,863,787. In contrast, Timor-Leste is the least damaged country with the total damages record of 13,636. Furthermore, Vietnam is the second most damaged country with 18,129,292 total records. Next to it are Indonesia, Malaysia, Philippines and Cambodia. There
are two countries with least equal total number damages: the Lao People and Myanmar.

```python
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20Deaths%20in%2020%20Years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python
#Visualization
#getting all values from the dataframe/csv
Country = df["Country"]
Deaths = df["Deaths"]

#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Creating plots
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("\n Flood Death Rate in Asean Country", fontweight="bold", fontsize=20)
plt.xlabel("Country", fontsize=15)
plt.ylabel("Death Count", fontsize=15)
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.plot(Country, Deaths, 'o-', lw=2.4, c='#333399')
plt.legend(labels=['Death Rate'])
plt.show()
```

```python
#getting all values from the dataframe/csv
Country = df["Country"]
Deaths = df["Deaths"]

#changing the size of the entire graph
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(7)

#Changing the background color of the chart
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#Assigning of various colors
colors = ['#488f31', '#7ba23e', '#a9b552', '#d5c86a', '#ffdc87', '#f9b76b', '#f19058', '#e56951', '#de425b']

#Adding a circle in the center for better visualization
center = plt.Circle((0,0), 0.5, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center)

#Creating Visualization
plt.pie(Deaths, labels=Country, colors = colors, autopct='%1.2f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, pctdistance=0.7, labeldistance=1.1, startangle=35, shadow = True, textprops={'fontsize': 13})
plt.legend()
plt.show()
```

```python
#Dataframe
url = 'https://raw.githubusercontent.com/mushimushe/Beta-Tech_BSIT_3207/main/Data%20Visualizations/Datasets%20csv/Total%20Floodsubtype%20in%2020%20Years.csv'
df = pd.read_csv(url)
display(df.iloc[0:])
```

```python

#getting all the values stored in the dataframe/csv
TypeofFlood = df["TypeofFlood"]
Frequency = df["Frequency"]

#changing the size of visualization
f = plt.figure(tight_layout=True)
f.set_figwidth(23)
f.set_figheight(10)

#Changing the background color of the graph
plt.rcParams['axes.facecolor'] = '#F2F2F2'

#creating the plot
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("\n Total Floodsubtype Incident in 20 Years", fontweight="bold", fontsize=20)
plt.xlabel("Type of Flood", fontsize=15)
plt.ylabel("Number of Incidents", fontsize=15)
plt.grid(color = 'white', linestyle = '--', linewidth = 1.5)
plt.plot(TypeofFlood, Frequency, color="#2f4b7c", lw=2.4, zorder=10)
plt.scatter(TypeofFlood, Frequency, fc="w", ec="#FF7A04", s=60, lw=2.4, zorder=12) 
#Saving visualizations in TIFF file with 300 dpi
plt.savefig('LaborMigration.tiff', transparent=True, dpi=300)
plt.show()
```
